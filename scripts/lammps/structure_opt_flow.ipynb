```python
from _lammps.get_thermo_from_log import get_thermo_csv_from_log
from _lammps.get_thermo_from_log import get_lattice_series_from_thermo_csv
from _lammps.structure_optimization import plot_opt
from _ovito.convert_dump_to_xsf import get_last_frame_xsf_file
from _ovito.compare_rdf import rdf_from_dump
from utils.constants.constants import Constants
import glob
import pandas as pd
import matplotlib.pyplot as plt

root_dir = '/Users/y1u0d2/desktop/Lab/result/lammps/structure-optimization/05'
structures = Constants.structures()
```

```python
# create thermo csv
for structure in structures:
    dirs = glob.glob(f'{root_dir}/{structure}/*')
    for _dir in dirs:
        get_thermo_csv_from_log(_dir, _dir)
```

```python
# create last structure xsf
for structure in structures:
    dirs = glob.glob(f'{root_dir}/{structure}/*.xsf')
    for _dir in dirs:
        print(_dir)
        get_last_frame_xsf_file(_dir)
```

```python
# create rdf.txt and rdf.csv from dump file
for structure in structures:
    dirs = glob.glob(f'{root_dir}/{structure}/*')
    for _dir in dirs:
        print(_dir)
        rdf_from_dump(
            path_to_dump=f'{_dir}/traj.dump',
            save_dir=_dir,
            rcut=15,
            bins=100
        )
```

```python
# create all lattice csv
for i, structure in enumerate(structures):
    dirs = glob.glob(f'{root_dir}/{structure}/*')
    for j, _dir in enumerate(dirs):
        print(_dir)
        if i==0 and j==0:
            df = get_lattice_series_from_thermo_csv(dirs[0])
            continue
        tmp_df = get_lattice_series_from_thermo_csv(_dir)
        df = pd.concat([df,tmp_df])
df.to_csv(f'{root_dir}/lattice.csv', index=False)
```

```python
# create all rdf csv
def format_rdf_df(rdf_df, idx, structure):
    df = rdf_df.T
    df.rename(index={'8-8':'O-O', '8-14':'O-Si', '14-14':'Si-Si'}, inplace=True)
    df.insert(0, 'idx', idx)
    df.insert(0, 'structure', structure)
    df.insert(0, 'bond', df.index)
    df.reset_index(drop=True, inplace=True)
    return df

for i,structure in enumerate(structures):
    dirs = glob.glob(f'{root_dir}/{structure}/*')
    for j,_dir in enumerate(dirs):
        structure = _dir.split('/')[-1].split('_')[1]
        idx = _dir.split('/')[-1].split('_')[-1].split('.')[0]
        rdf_df = pd.read_csv(f'{_dir}/rdf.csv', index_col='distance')
        if i==0 and j==0:
            df = format_rdf_df(rdf_df, idx=idx, structure=structure)
            continue
        tmp_df = format_rdf_df(rdf_df, idx=idx, structure=structure)
        df = pd.concat([df,tmp_df])
df.reset_index(drop=True, inplace=True)
df.to_csv(f'{root_dir}/rdf.csv', index=False)
```

```python
# plot rdf comparison
def get_stable_structure_csv(structure):
    files = glob.glob(f'/Users/y1u0d2/desktop/Lab/data/stable-structure/rdf_csv/*')
    file = list(filter(lambda file: structure in file, files))
    if len(file) != 1:
        return
    return file[0]

def plot_rdf_comparison(df_obs, df_pred, structure, bond_num, bond, save_dir):
    fig, ax = plt.subplots()
    ax.set_title(f'comparison of RDF: {structure} {bond}')
    ax.set_xlabel('r(Å)')
    ax.set_ylabel('g(r)')
    ax.plot(df_pred.distance, df_pred[bond_num], label='pred')
    ax.plot(df_obs.distance, df_obs[bond], label='obs')
    ax.legend()
    fig.savefig(f'{save_dir}/{structure}_{bond}.png')

structures = Constants.structures()
for structure in structures:
    obs_dir = get_stable_structure_csv(structure)
    dirs = glob.glob(f'{root_dir}/{structure}/*')
    for pred_dir in dirs:
        df_pred = pd.read_csv(f'{pred_dir}/rdf.csv')
        df_obs = pd.read_csv(obs_dir)
        plot_rdf_comparison(df_obs, df_pred, structure, '8-8', 'O-O', pred_dir)
        plot_rdf_comparison(df_obs, df_pred, structure, '8-14', 'O-Si', pred_dir)
        plot_rdf_comparison(df_obs, df_pred, structure, '14-14', 'Si-Si', pred_dir)
```

```python
# plot structure optimization trajectory
for structure in structures:
    dirs = glob.glob(f'{root_dir}/{structure}/*')
    for _dir in dirs:
        csv = glob.glob(f'{_dir}/thermo_*.csv')[0]
        natom = csv.split('/')[-1].split('.')[0].split('_')[-1]
        plot_opt(_dir, structure, int(natom),savename=f'{structure}_min', is_min=True)
```
