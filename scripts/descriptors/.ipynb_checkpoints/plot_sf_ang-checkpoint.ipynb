```python
import matplotlib.pyplot as plt
plt.style.use({'figure.facecolor':'white'})
import pandas as pd
import numpy as np
```

```python
def ang_symmetry_function_3(theta, lambdas, zeta, is_degree=True):
    if is_degree:
        theta = np.radians(theta)
    return 2**(1-zeta) * (1 + lambdas * np.cos(theta))**zeta
```

```python
params = [
    [-1, 1.000E+00],
    [ 1, 1.000E+00],
    [-1, 6.000E+00],
    [ 1, 6.000E+00],
    # [-1, 1.000E+00],
    # [ 1, 1.000E+00],
    # [-1, 6.000E+00],
    # [ 1, 6.000E+00],
]
x = np.linspace(0, 360)
fig, ax = plt.subplots(figsize=(8,6))
for param in params:
    lambdas = param[0]
    zeta = param[1]
    y = ang_symmetry_function_3(theta=x, lambdas=lambdas, zeta=zeta)
    ax.set_xlim(0, 360)
    ax.set_xlabel('degree/°')
    ax.set_ylabel(r'$2^{1-\zeta} (1+\lambda \cos\theta_{ijk})^\zeta$')
    ax.plot(x, y, label=f'lambda:{lambdas}, zeta:{zeta}')
    ax.legend()
fig.savefig('/Users/y1u0d2/desktop/Lab/result/sf/ang/ang_2.png')
```

```python
def _soft_plus(x):
    return np.log(1 + np.exp(x))

x = np.linspace(-5,5)
y = _soft_plus(x)
# plt.plot(x,np.zeros(x))
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0,6)
plt.plot(x,x, label="y=x")
plt.plot(x,y, label="softplus")
plt.legend()
```
