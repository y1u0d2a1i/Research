```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use({'figure.facecolor':'white'})
from utils.constants.dir_path import DataDirPath
```

```python
base_df = pd.read_csv(f'{DataDirPath.base_structure_info()}/base_info_convert.csv')
# base_df = pd.read_csv(f'{DataDirPath.base_structure_info()}/base_info_convert.csv')
```

```python
base_df['Esio2'] = base_df.E / (base_df.natom/3)
```

```python
base_df['Vsio2'] = base_df.Vol / (base_df.natom/3)
```

```python
min_e = base_df.Esio2.min()
base_df['Ediff'] = base_df.Esio2 - min_e
```

```python
structures = base_df.structure.unique()
for structure in structures:
    tmp = base_df.copy()
    df = tmp[tmp.structure == structure]
    plt.title(f'{structure}')
    plt.xlabel('Vol(ang^3/SiO2)')
    plt.ylabel('E(eV/SiO2)')
    plt.scatter(df.Vsio2, df.Ediff)
    plt.show()
```
